```python
#Removing Outliers

def remove_outliers(house_set, column):
    Q1 = house_set[column].quantile(0.25)
    Q3 = house_set[column].quantile(0.75)
    
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    return house_set[(house_set[column] >= lower_bound) & (house_set[column] <= upper_bound)]
    
x=["price","bedrooms","bathrooms","sqft_living","sqft_lot"]

for i in x:
    house_set = remove_outliers(house_set, i)
```

```python
# Feature Engineering
house_set['house_age'] = 2025 - house_set['yr_built']
house_set['renovated_age'] = house_set['yr_renovated'].apply(lambda x: 2025 - x if x > 0 else 0)
house_set['price_per_sqft'] = house_set['price'] / house_set['sqft_living']
house_set['total_sqft'] = house_set['sqft_living'] + house_set['sqft_basement']
house_set['bath_bed_ratio'] = house_set['bathrooms'] / house_set['bedrooms'].replace(0, 1)
house_set.dropna(inplace=True)

house_set.head()
```

```python
from scipy import stats

# T-test: Waterfront vs Non-Waterfront
price_waterfront = house_set[house_set['waterfront'] == 1]['price']
price_non_waterfront = house_set[house_set['waterfront'] == 0]['price']

t_stat, p_val_ttest = stats.ttest_ind(price_waterfront, price_non_waterfront, equal_var=False)

print("T-test: Waterfront vs Non-Waterfront")
print(f"T-statistic = {t_stat:.3f}, p-value = {p_val_ttest:.5f}\n")


alpha = 0.05
if p_val_ttest < alpha:
    print("Result: Significant difference in price between waterfront and non-waterfront properties.")
else:
    print("Result: No significant difference in price between waterfront and non-waterfront properties.")


print("\n--------------------------------------------------------------------------------------------------n") 


# ANOVA: Price by Number of Bedrooms
grouped_prices = [group['price'].values for name, group in house_set.groupby('bedrooms')]
f_stat, p_val_anova = stats.f_oneway(*grouped_prices)

print("ANOVA: Price by Number of Bedrooms")
print(f"F-statistic = {f_stat:.3f}, p-value = {p_val_anova:.5e}\n")

# Interpretation of ANOVA
if p_val_anova < alpha:
    print("Result: At least one bedroom group has a significantly different mean price.")
else:
    print("Result: No significant difference in mean prices across bedroom groups.")
    
    
```
